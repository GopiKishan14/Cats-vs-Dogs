````python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
````

````python
DATADIR = './PetImages'  #currently in .gitignore , link to dataset given in README
CATEGORIES = ["Dog" , "Cat"]
````

````python
# Visualising and eliminating errors for one data.
for category in CATEGORIES:
    path = os.path.join(DATADIR , category) # creating path to 'Dog' and 'Cat' dir
    for img in os.listdir(path):  #iterate over all images in Dog and Cat
        img_arr = cv2.imread(os.path.join(path , img) , cv2.IMREAD_GRAYSCALE) #convert to array
        plt.imshow(img_arr , cmap='gray') #grap it
        plt.show() #display it
        
        break
    break
        
        
````

````python
print(img_arr.shape , img_arr)
````

````python
IMG_SIZE = 50
new_arr = cv2.resize(img_arr , (IMG_SIZE ,IMG_SIZE))
plt.imshow(new_arr , cmap='gray')
plt.show()
````

````python
# Actually creating training data. Almost same code as above.
training_data = []

def create_training_data():
    for category in CATEGORIES:
        path = os.path.join(DATADIR , category)
        class_num = CATEGORIES.index(category)  # get the classification  (0 or a 1). 0=dog 1=cat
        
        for img in os.listdir(path):
            try:
                img_arr = cv2.imread(os.path.join(path , img), cv2.IMREAD_GRAYSCALE)
                new_arr = cv2.resize(img_arr , (IMG_SIZE , IMG_SIZE))
                training_data.append([new_arr , class_num])
                
            except Exception as e:
                pass
            
create_training_data() #calling the function
print(len(training_data))
````

````python
import random
random.shuffle(training_data)
````

````python
for sample in training_data[:10]:
    print(sample[1])
````

````python
X = [] 
Y = []

for features , label in training_data:
    X.append(features)
    Y.append(label)

X = np.array(X).reshape(-1 , IMG_SIZE ,IMG_SIZE ,1)
````

````python
print(X.shape , len(Y)) #note X is arr and Y is list
````

````python
import pickle

pickle_out = open("X_50.pickle" , "wb")
pickle.dump(X , pickle_out)
pickle_out.close()

pickle_out = open("Y_50.pickle" , "wb")
pickle.dump(Y , pickle_out)
pickle_out.close()
````
